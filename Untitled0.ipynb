Preparando entorno

```python
# Use seaborn for pairplot
!pip install -q seaborn
import pathlib
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers

print(tf.__version__)
```

Cargando datos

```python
dataset_path = keras.utils.get_file("auto-mpg.data", "http://archive.ics.uci.edu/ml/machine-learning-databases/auto-mpg/auto-mpg.data")
dataset_path
column_names = ['mpg','cilindros','desplazamiento','horsepower','peso','aceleracion','modelo','origen']
raw_dataset = pd.read_csv(dataset_path,names=column_names,na_values="?",comment='\t',sep=" ",skipinitialspace=True)
dataset = raw_dataset.copy()
type(dataset)
```

```python
hola = {'columna_1':['uno_uno','uno_dos','uno_tres'],'columna_2':['dos_uno','dos_dos','dos_tres']}
df = pd.DataFrame(data=hola)
df
# df.at[2,'columna_1'] # elemento muy particular
# df.iloc[1] # elementos de una fila
# dataset[:10] # 10 filas
# df.size
# df['columna']
mi_pop = df.pop('columna_1')
type(mi_pop)
```

Preparando datos

```python
dataset = dataset.dropna()
origin = dataset.pop('origen')
dataset['USA'] = (origin == 1)*1.0
dataset['Europe'] = (origin == 2)*1.0
dataset['Japan'] = (origin == 3)*1.0

train_dataset = dataset.sample(frac=0.8,random_state=0)
test_dataset = dataset.drop(train_dataset.index)

train_stats = train_dataset.describe()

train_labels = train_dataset.pop('mpg')
test_labels  = test_dataset.pop('mpg')
```

```python
dataset
```

Normalizando datos

```python
def norm(x):
  return (x - train_stats['mean']) / train_stats['std']
normed_train_data = norm(train_dataset)
normed_test_data = norm(test_dataset)
normed_train_data
```

modelo

```python
def build_model():
  model = keras.Sequential([
    layers.Dense(64, activation='relu', input_shape=[len(train_dataset.keys())]),
    layers.Dense(64, activation='relu'),
    layers.Dense(1)
  ])

  optimizer = tf.keras.optimizers.RMSprop(0.001)

  model.compile(loss='mse',
                optimizer=optimizer,
                metrics=['mae', 'mse'])
  return model

model = build_model()
model.summary()
```

```python
example_batch = normed_train_data[:10]
example_result = model.predict(example_batch)
example_result
```

```python
# Display training progress by printing a single dot for each completed epoch
class PrintDot(keras.callbacks.Callback):
  def on_epoch_end(self, epoch, logs):
    if epoch % 100 == 0: print('')
    print('.', end='')

EPOCHS = 1000

history = model.fit(
  normed_train_data, 
  train_labels,
  epochs=EPOCHS, 
  validation_split = 0.2, 
  verbose=0,
  callbacks=[PrintDot()])
```
